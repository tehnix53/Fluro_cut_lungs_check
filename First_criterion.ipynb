```python
import os
import numpy as np
from PIL import Image
from matplotlib.pyplot import imshow
```

```python
work_dir = '/home/alex/Документы/pres_fasie/технический отчет/first_criterion/'
```

```python
def show_image(name):
    sample = np.asarray(Image.open(work_dir+name))
    sample = np.where(sample>0,255,0)
    imshow(sample)
```

```python
def Percentage(number,percent):
    return int(number*percent/100)    
```

```python
def first_criterion(image,treshold):
    array = np.asarray(Image.open(image))
    array = np.where(array > 0,255,0)      
    size = array.shape    
    treshold = Percentage(size[0], treshold)
    
    top = 0
    bottom = 0
    left = 0
    right = 0
    
    # top
    for i in range (treshold):
        if np.amax(array[i]) > 1:
            top = 1
            break
        
    # bottom
    for i in range (size[0]-treshold,size[0]):
        if np.amax(array[i])>1:
            bottom = 1
            break
    
    # left
    for i in array:
        if np.amax(i[0:treshold])>1:
            left = 1
            break
            
    # right
    for i in array:
        if np.amax(i[size[1]-treshold:size[1]]) > 1:
            right = 1
            break
        
    return [top,bottom,left,right]
```

```python
samples = [i for i in os.listdir(work_dir) if i.endswith('.png')]
```

```python
image = samples[0]
show_image(image)
first_criterion(work_dir+image,5)
```
